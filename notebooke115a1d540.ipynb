```python
!ffmpeg -y \
    -i /kaggle/input/datasets/sudhanva7/supernan/clip.mp4 \
    -i /kaggle/working/hindi_delayed.wav \
    -c:v libx264 \
    -c:a aac \
    -ar 44100 \
    -map 0:v:0 \
    -map 1:a:0 \
    -shortest \
    /kaggle/working/output_dubbed.mp4

size = os.path.getsize('/kaggle/working/output_dubbed.mp4')/1024/1024
print(f"‚úÖ Final video: {size:.1f} MB")
```

```python
import gdown
import os

gdown.download(
    "https://drive.google.com/file/d/1oe3roQCGEebUf6NlRvdWrh6603dIkZKg/view?usp=sharing",
    "/kaggle/working/input_video.mp4",
    quiet=False,
    fuzzy=True
)

size = os.path.getsize('/kaggle/working/input_video.mp4')/1024/1024
print(f"‚úÖ Downloaded: {size:.1f} MB")
```

```python
# Split into 3 parts, slow middle, rejoin
# Part 1: 0-3 seconds (normal)
!ffmpeg -y -i /kaggle/working/input_video.mp4 \
    -ss 00:00:00 -to 00:00:03 \
    -c:v libx264 -c:a aac \
    /kaggle/working/part1.mp4

# Part 2: 3-9 seconds (slowed to 0.6x speed)
!ffmpeg -y -i /kaggle/working/input_video.mp4 \
    -ss 00:00:03 -to 00:00:09 \
    -vf "setpts=1.5*PTS" \
    -af "atempo=0.667" \
    /kaggle/working/part2.mp4

# Part 3: 9-20 seconds (normal)
!ffmpeg -y -i /kaggle/working/input_video.mp4 \
    -ss 00:00:09 -to 00:00:20 \
    -c:v libx264 -c:a aac \
    /kaggle/working/part3.mp4

print("‚úÖ All parts created!")
```

```python
# Create file list
with open('/kaggle/working/filelist.txt', 'w') as f:
    f.write("file '/kaggle/working/part1.mp4'\n")
    f.write("file '/kaggle/working/part2.mp4'\n")
    f.write("file '/kaggle/working/part3.mp4'\n")

# Merge
!ffmpeg -y -f concat -safe 0 \
    -i /kaggle/working/filelist.txt \
    -c:v libx264 -c:a aac \
    /kaggle/working/output_final.mp4

import os
size = os.path.getsize('/kaggle/working/output_final.mp4')/1024/1024
print(f"‚úÖ Final video: {size:.1f} MB")
```
